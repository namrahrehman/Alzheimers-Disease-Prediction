```r
# Load and inspect the dataset
data <- read.csv("/kaggle/input/alzheimers-disease-dataset/alzheimers_disease_data.csv")
head(data)
str(data)
summary(data)
colSums(is.na(data))

# Visualize Age Distribution
ggplot(data, aes(x = Age)) + 
  geom_histogram(binwidth = 1, fill = "blue", color = "black") + 
  theme_minimal() + 
  ggtitle("Age Distribution")

# Visualize MMSE Distribution
ggplot(data, aes(x = MMSE)) + 
  geom_histogram(binwidth = 1, fill = "green", color = "black") + 
  theme_minimal() + 
  ggtitle("MMSE Distribution")

# Visualize Gender Distribution
ggplot(data, aes(x = as.factor(Gender))) + 
  geom_bar(fill = "purple", color = "black") + 
  theme_minimal() + 
  ggtitle("Gender Distribution")
```

```r
# Feature Engineering
data$lifestyle_score <- data$BMI * data$PhysicalActivity
data$Gender <- as.factor(data$Gender)
data$Ethnicity <- as.factor(data$Ethnicity)
data$age_family_interaction <- data$Age * data$FamilyHistoryAlzheimers
data$Log_CholesterolTotal <- log(data$CholesterolTotal + 1)

# Split the data (80% training, 20% testing)
set.seed(123)
train_index <- createDataPartition(data$Diagnosis, p = 0.8, list = FALSE)
train_data <- data[train_index, ]
test_data <- data[-train_index, ]
```

```r
# Correlation matrix
cor_matrix <- cor(train_data[, sapply(train_data, is.numeric)])
print(cor_matrix)

# Statistical Tests
# T-test to compare Age between Diagnosis groups
t_test_age <- t.test(Age ~ Diagnosis, data = train_data)
print(t_test_age)

# Chi-squared test for Gender and Diagnosis
chi_squared_gender <- chisq.test(table(train_data$Gender, train_data$Diagnosis))
print(chi_squared_gender)

# T-test for BMI between Diagnosis groups
t_test_bmi <- t.test(BMI ~ Diagnosis, data = train_data)
print(t_test_bmi)

# T-test for MMSE
t_test_mmse <- t.test(MMSE ~ Diagnosis, data = train_data)
print(t_test_mmse)

# T-test for Cholesterol Total
t_test_cholesterol_total <- t.test(CholesterolTotal ~ Diagnosis, data = train_data)
print(t_test_cholesterol_total)

# Chi-squared tests for categorical variables
chi_squared_family_history <- chisq.test(table(train_data$FamilyHistoryAlzheimers, train_data$Diagnosis))
print(chi_squared_family_history)

chi_squared_diabetes <- chisq.test(table(train_data$Diabetes, train_data$Diagnosis))
print(chi_squared_diabetes)

chi_squared_cvd <- chisq.test(table(train_data$CardiovascularDisease, train_data$Diagnosis))
print(chi_squared_cvd)

chi_squared_depression <- chisq.test(table(train_data$Depression, train_data$Diagnosis))
print(chi_squared_depression)
```

```r
# %% [code]
# Fit logistic regression model
log_model <- glm(Diagnosis ~ ., data = train_data, family = "binomial")

# Make predictions on the test set
predictions <- predict(log_model, newdata = test_data, type = "response")
predicted_class <- ifelse(predictions > 0.5, 1, 0)  # Threshold = 0.5

# Confusion matrix
conf_matrix <- confusionMatrix(factor(predicted_class), factor(test_data$Diagnosis))
print(conf_matrix)

# ROC Curve
library(pROC)
roc_curve <- roc(test_data$Diagnosis, predictions)
plot(roc_curve, main = "ROC Curve for Logistic Regression Model")

# %% [code]
# Prepare the data
train_data_matrix <- as.matrix(train_data[, -which(names(train_data) == "Diagnosis")])  # Remove the target column
train_labels <- train_data$Diagnosis  # The target variable

# Lasso Regularization (L1)
logreg_lasso <- glmnet(train_data_matrix, train_labels, family = "binomial", alpha = 1, lambda = 0.1)
summary(logreg_lasso)

# Ridge Regularization (L2)
logreg_ridge <- glmnet(train_data_matrix, train_labels, family = "binomial", alpha = 0, lambda = 0.1)
summary(logreg_ridge)

# %% [code]
# Coefficients for Lasso (L1)
coef(logreg_lasso)

# Coefficients for Ridge (L2)
coef(logreg_ridge)

# %% [code]
# Deviance for Lasso (L1)
logreg_lasso$dev.ratio

# Deviance for Ridge (L2)
logreg_ridge$dev.ratio

# %% [code]
# Plot regularization path for Lasso (L1)
plot(logreg_lasso, xvar = "lambda", label = TRUE)

# Plot regularization path for Ridge (L2)
plot(logreg_ridge, xvar = "lambda", label = TRUE)

# %% [code]
# Cross-validation for Lasso (L1)
cv_lasso <- cv.glmnet(train_data_matrix, train_labels, family = "binomial", alpha = 1)
plot(cv_lasso)
best_lambda_lasso <- cv_lasso$lambda.min

# Cross-validation for Ridge (L2)
cv_ridge <- cv.glmnet(train_data_matrix, train_labels, family = "binomial", alpha = 0)
plot(cv_ridge)
best_lambda_ridge <- cv_ridge$lambda.min

# Refit models with the best lambda
logreg_lasso_best <- glmnet(train_data_matrix, train_labels, family = "binomial", alpha = 1, lambda = best_lambda_lasso)
logreg_ridge_best <- glmnet(train_data_matrix, train_labels, family = "binomial", alpha = 0, lambda = best_lambda_ridge)

# %% [code]
# Coefficients for Lasso (L1)
coef(logreg_lasso_best)

# Coefficients for Ridge (L2)
coef(logreg_ridge_best)

# %% [markdown]
# Lasso (L1) and Ridge (L2) Regularization Results:
# 
# **Lasso (L1) Regularization:**
# - Feature selection was performed, excluding Age and Gender.
# - Positive coefficient for BMI, suggesting a slight increase in Alzheimer's likelihood with higher BMI.
# - Negative coefficient for MMSE, indicating that lower MMSE scores increase the likelihood of Alzheimer's.
# - FamilyHistoryAlzheimers and EducationLevel: Negative coefficients suggesting their association with higher Alzheimer's likelihood.
# 
# **Ridge (L2) Regularization:**
# - Retained all features but reduced the magnitude of coefficients.
# - Age: Small negative coefficient, weak inverse relationship with Alzheimer's.
# - BMI: Positive coefficient, weak positive relationship with Alzheimer's.
# - MMSE: Negative coefficient, consistent with Lasso.
# - FamilyHistoryAlzheimers, Gender, EducationLevel: Negative relationships with Alzheimer's, with FamilyHistoryAlzheimers showing the strongest negative association.

# %% [markdown]
# Cross-validation for Model Evaluation

# %% [code]
library(caret)

# Define cross-validation settings (5-fold cross-validation)
train_control <- trainControl(method = "cv", number = 5)

# Lasso Model Cross-validation
cv_lasso <- train(Diagnosis ~ ., data = train_data, method = "glmnet", 
                  family = "binomial", trControl = train_control, tuneGrid = expand.grid(alpha = 1, lambda = seq(0.01, 0.1, length = 10)))

# Ridge Model Cross-validation
cv_ridge <- train(Diagnosis ~ ., data = train_data, method = "glmnet", 
                  family = "binomial", trControl = train_control, tuneGrid = expand.grid(alpha = 0, lambda = seq(0.01, 0.1, length = 10)))

# Show results
print(cv_lasso)
print(cv_ridge)
```

```r
# Load the randomForest library
library(randomForest)

# Train a random forest model
rf_model <- randomForest(Diagnosis ~ ., data = train_data, ntree = 500, mtry = 3, importance = TRUE)

# Print the model summary
print(rf_model)

# Make predictions on the test set
predictions_rf <- predict(rf_model, test_data)

# Create confusion matrix
conf_matrix_rf <- table(test_data$Diagnosis, predictions_rf)

# Calculate accuracy, precision, recall, and F1-score
accuracy_rf <- sum(diag(conf_matrix_rf)) / sum(conf_matrix_rf)
precision_rf <- conf_matrix_rf[2,2] / sum(conf_matrix_rf[,2])
recall_rf <- conf_matrix_rf[2,2] / sum(conf_matrix_rf[2,])
f1_rf <- 2 * ((precision_rf * recall_rf) / (precision_rf + recall_rf))

# Print the evaluation metrics
cat("Random Forest Model Performance:\n")
cat("Accuracy: ", accuracy_rf, "\n")
cat("Precision: ", precision_rf, "\n")
cat("Recall: ", recall_rf, "\n")
cat("F1-Score: ", f1_rf, "\n")
```
